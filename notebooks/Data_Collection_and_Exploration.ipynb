```python
### Doenload Data for Multiple Companies

Comapny_Name = ['SYK', 'MDT', 'BSX','JNJ']
multi_data = yf.download(Comapny_Name, start='2023-01-01', end='2025-04-30')
```

```python
multi_data.tail()
```

```python
multi_data['Close'].plot(figsize=(12,6), title='Closing Prices of Med-Tech Stocks' ,)
plt.xlabel('Date')
plt.ylabel('Prices in $')
```

```python
# Mapping of ticker symbols to full company names
stock_names = {
    'BSX': 'Boston Scientific',
    'JNJ': 'Johnson & Johnson',
    'MDT': 'Medtronic',
    'SYK': 'Stryker'
}
# Rename columns
multi_data_renamed = multi_data.rename(columns=stock_names)

# Plot with full names
multi_data_renamed['Close'].plot(figsize=(12, 6), title='Closing Prices of Med-Tech Stocks')
plt.xlabel('Date')
plt.ylabel('Prices in $')
```

```python
# Plot with full names
Stryker_close_data=multi_data_renamed['Close']['Stryker']
Stryker_close_data.tail()
```

```python
Stryker_pct_change_data= Stryker_close_data.pct_change()*100
Stryker_pct_change_data.tail()
```

```python
Stryker_pct_change_data.dropna().plot(figsize=(12, 6), title='Daily % Change in Closing Prices of Stryker Stock')
```

```python
close_data = multi_data.rename(columns=stock_names)['Close']
# Step 3: Compute daily % change
pct_change_data = close_data.pct_change() * 100

# Step 4: Plot
ax = pct_change_data.plot(figsize=(18, 10), title='Daily % Change in Closing Prices of Med-Tech Stocks')
plt.ylabel('Daily % Change')

# Step 5: Highlight max gain and loss for each stock
for stock in pct_change_data.columns:
    max_idx = pct_change_data[stock].idxmax()
    min_idx = pct_change_data[stock].idxmin()
    
    # Plot points
    ax.plot(max_idx, pct_change_data.loc[max_idx, stock], 'g^', markersize=10, label=f'{stock} Max Gain')
    ax.plot(min_idx, pct_change_data.loc[min_idx, stock], 'rv', markersize=10, label=f'{stock} Max Loss')
    
    # Annotate
    ax.annotate(f'{pct_change_data.loc[max_idx, stock]:.2f}%', 
                (max_idx, pct_change_data.loc[max_idx, stock]), 
                textcoords="offset points", xytext=(0,10), ha='center', color='green')
    
    ax.annotate(f'{pct_change_data.loc[min_idx, stock]:.2f}%', 
                (min_idx, pct_change_data.loc[min_idx, stock]), 
                textcoords="offset points", xytext=(0,-15), ha='center', color='red')

# Optional: clean up duplicate labels in legend
handles, labels = ax.get_legend_handles_labels()
unique_labels = dict(zip(labels, handles))
ax.legend(unique_labels.values(), unique_labels.keys())

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Calculate Returns

```python
# Daily Returns
stryker_data['Daily Return'] = stryker_data['Close'].pct_change()
stryker_data[-100:]['Daily Return'].plot(figsize=(12,6), title = " Daily Return - Stryker")
```

```python
# Cumulative Return

stryker_data['Cumulative Return'] = (1+ stryker_data['Daily Return']).cumprod()
stryker_data[-100:]['Cumulative Return'].plot(figsize=(12,6), title = 'Cumlative Return - Stryker')
```

```python
# Log Returns (more advanced)
import numpy as np
stryker_data['Log Return'] = np.log(stryker_data['Close'] / stryker_data['Close'].shift(1))
stryker_data[-100:]['Log Return'].plot(figsize=(12,6), title = 'Log Return - Stryker')
```

# Data Collector Function

```python
'''
Data_Collector_Function
    - Input Arguments - Ticker Symbol(company Name) , Start , end
    - It will download and export data to csv and store it data dir. 
    - For Apple = StockDataCollector('AAPL',start='2024-04-01', end='2025-04-30')
'''
def StockDataCollector(ticker_symbol, start , end):
    import yfinance as yf
    import pandas as pd
    try:
        stock_data = yf.download(ticker_symbol,start=start,end=end)
        #ticker = yf.Ticker(ticker_symbol)
        #company_name= ticker.info['longName']
        stock_data.to_csv(f'../data/{ticker_symbol}_{start}_to_{end}.csv',index=True)
        return 1
    except:
        print("An exception occurred")
        return 0
```

```python
# Download reliance data - RELIANCE.BO

status = StockDataCollector("^NSEI",'2023-04-01','2025-05-06')
if status ==1 :
    print("Data Downloaded Sucessfully")
else:
    print("An exception occurred")
```

```python
# Download reliance data - RELIANCE.BO

status = StockDataCollector("SYK",'2019-04-01','2025-05-06')
if status ==1 :
    print("Data Downloaded Sucessfully")
else:
    print("An exception occurred")
```
