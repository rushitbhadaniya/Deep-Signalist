## Create Reusable Functions

### 1. News API KEY Loader

```python
def get_news_api():
    """
    Loads the NewsAPI key from a JSON configuration file.

    Returns:
        str: The NewsAPI key used for making authenticated requests.
    """
    import json
    # Open the configuration file and load the contents
    with open("..\\config.json") as config_file:
        config = json.load(config_file)

    # Retrieve the NewsAPI key from the configuration
    NEWS_API_KEY = config["NEWSAPI_KEY"]
    
    # Return the API key
    return NEWS_API_KEY
```

### 2. News Fatcher 

```python
import requests
import pandas as pd
from datetime import datetime
import os

def get_news(ticker :str , from_date:str , to_date:str , save:bool = True) -> pd.DataFrame:
    """
    Fetches financial news articles for a given ticker/company using NewsAPI.org
    Args:
        ticker (str): Company name or stock ticker
        from_date (str): 'YYYY-MM-DD'
        to_date (str): 'YYYY-MM-DD'
        save (bool): If True, saves results to CSV
    
    Returns:
        pd.DataFrame: News articles dataframe
    """
    # Define the target directory path
    output_dir = '..\\data\\news_data'

    # Create the directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    API_KEY = get_news_api()
    query = f"{ticker} stock"
    url = (
        f"https://newsapi.org/v2/everything?"
        f"q={query}&from={from_date}&to={to_date}&language=en&"
        f"sortBy=publishedAt&pageSize=100&apiKey={API_KEY}"
    )
    
    response = requests.get(url)
    data = response.json() 
    articles = data.get('articles', [])
    
    # No Articles Founds
    if not articles:
        print(f"No articles found for {ticker} from {from_date} to {to_date}.")
        return pd.DataFrame()
    
    news_dataframe = pd.DataFrame([{
        'ticker': ticker.upper(),
        'source':article['source']['name'],
        'title': article['title'],
        'description': article['description'],
        'content': article['content'],
        'url': article['url'],
        'publishedAt': pd.to_datetime(article['publishedAt'])
        
    } for article in articles])
    
    if save:
        filename = f"{ticker.upper()}_{from_date}_to_{to_date}.csv"
        news_dataframe.to_csv(f'{output_dir}/{filename}', index=False)
        print(f"✅ Saved: '{output_dir}/{filename} ")

    return news_dataframe
```

```python
tickers = ['AAPL', 'TSLA', 'GOOGL', 'MSFT']
from_date = '2025-05-10'
to_date = '2025-05-15'

all_news = []

for ticker in tickers:
    df = get_news(ticker, from_date, to_date)
    if not df.empty:
        all_news.append(df)

# Combine all
if all_news:
    combined_news = pd.concat(all_news, ignore_index=True)
    combined_news.to_csv(f'../data/news_data/Combined_News_MultiTickers.csv', index=False)
    print(f"✅ Saved: ../data/news_data/Combined_News_MultiTickers.csv")
```

### 3. News Pipeline

```python
def run_news_pipeline(ticker_list, from_date, to_date, combine: bool = True):
    all_news = []
    
    for ticker in tickers:
        df = get_news(ticker, from_date, to_date)
        if not df.empty:
            all_news.append(df)
    
    # Combine all
    if combine and all_news:
        combined_news = pd.concat(all_news, ignore_index=True)
        combined_news.to_csv(f'../data/news_data/Combined_News_MultiTickers.csv', index=False)
        print(f"✅ Saved: ../data/news_data/Combined_News_MultiTickers.csv")
```

```python
tickers = ["AAPL", "TSLA", "MSFT", "GOOGL"]
from_date = "2025-05-10"
to_date = "2025-05-15"

run_news_pipeline(ticker_list=tickers, from_date=from_date, to_date=to_date)
```
